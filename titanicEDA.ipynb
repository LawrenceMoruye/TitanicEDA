```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
df_train=pd.read_csv("train.csv")
```

```python
for x in [1,2,3]:
    df_train.Survived[df_train.Pclass==x].plot(kind="kde")
plt.legend(("1st","2nd","3rd"))
plt.title("survived wrt pclass")
plt.show()
```

```python
df_train.Survived[(df_train.Sex=="male")&(df_train.Pclass==1)].value_counts(normalize=True).plot(kind="bar")
plt.title("rich men")
plt.show()
```

```python
df_train.Survived[(df_train.Sex=="female")&(df_train.Pclass==1)].value_counts(normalize=True).plot(kind="bar")
plt.title("rich women")
plt.show()
```
